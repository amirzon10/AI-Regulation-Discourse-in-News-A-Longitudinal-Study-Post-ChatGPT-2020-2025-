# The fuction of this notebook is to create a final database that can be used acroosed notebooks for diverse analysis

```python
# Import libraries
import pandas as pd 
```

```python
# let's load both dataset
df1 = pd.read_csv("../data/media_cloud-political_orientation.csv")
df2 = pd.read_csv("../data/mediacloud-translated-complete.csv")

# I want to add a column to df2 that containes political orientation from df1
# but not all instances in df1 are in df2, but we have id column in both datasets that can help us merge them

df = pd.merge(df2, df1[['id', 'political_orientation']], on='id', how='left')

print(f"df1 shape: {df1.shape}")
```

```python
print(f"df2 shape: {df2.shape}")
```

```python
print(f"df shape: {df.shape}")
```

```python
df.to_csv("../data/mediacloud-PO-translation.csv", index=False)
```

we run use this dataset to do sentiment analysis

```python
df = pd.read_csv("../data/mediacloud-PO-translation.csv")
```

```python
#let's now marge in the same way a new column for a df3
df3 = pd.read_csv("../data/mediacloud-sentiment-analysis.csv")
df = pd.merge(df, df3[['id', 'VADER_Compound_Score',
       'Sentiment_VADER']], on='id', how='left')
```
